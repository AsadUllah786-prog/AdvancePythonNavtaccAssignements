```python
# Initial Setup
user_name = "Asad"
correct_pin = 1234
balance = 5000
min_balance = 1000
transaction_count = 0
max_attempts = 3

print("===== Welcome to Smart ATM =====")

# Login System
attempts = 0

while attempts < max_attempts:
    name = input("Enter your name: ")
    pin = int(input("Enter your PIN: "))

    if name == user_name and pin == correct_pin:
        print("Login successful!\n")
        break
    else:
        attempts += 1
        print("Incorrect name or PIN.")
        print(f"Attempts left: {max_attempts - attempts}\n")

if attempts == max_attempts:
    print("Too many failed attempts. Account locked.")
    exit()

# ATM Menu Loop
while True:
    print("\n===== ATM MENU =====")
    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Exit")

    choice = input("Choose an option (1-4): ")

    if choice == "1":
        print(f"Current Balance: {balance}")

    elif choice == "2":
        amount = float(input("Enter deposit amount: "))

        if amount > 0:
            balance += amount
            transaction_count += 1
            print(f"Deposit successful. New Balance: {balance}")
        else:
            print("Deposit amount must be positive.")

    elif choice == "3":
        amount = float(input("Enter withdraw amount: "))

        if amount <= 0:
            print("Withdrawal amount must be positive.")
        elif balance - amount < min_balance:
            print("Cannot withdraw. Minimum balance must remain 1000.")
        else:
            balance -= amount
            transaction_count += 1
            print(f"Withdrawal successful. New Balance: {balance}")

    elif choice == "4":
        print("\nThank you for using Smart ATM.")
        print(f"Total transactions made: {transaction_count}")
        break

    else:
        print("Invalid option. Please choose between 1-4.")
```
