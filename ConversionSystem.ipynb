```python

print("Unit Conversion System")
print("Categories: 1. Length  2. Temperature  3. Weight  4. Time")

choice = input("Select a category (1-4): ")

value = float(input("Enter the value to convert: "))
from_unit = input("Enter the current unit: ").lower()
to_unit = input("Enter the unit to convert into: ").lower()

result = None

# Length conversions
if choice == "1":
    # Convert to meters first
    if from_unit == "meters":
        meters = value
    elif from_unit == "kilometers":
        meters = value * 1000
    elif from_unit == "inches":
        meters = value * 0.0254
    elif from_unit == "feet":
        meters = value * 0.3048
    else:
        meters = None

    if meters is not None:
        if to_unit == "meters":
            result = meters
        elif to_unit == "kilometers":
            result = meters / 1000
        elif to_unit == "inches":
            result = meters / 0.0254
        elif to_unit == "feet":
            result = meters / 0.3048

# Temperature conversions
elif choice == "2":
    if from_unit == "celsius":
        if to_unit == "fahrenheit":
            result = (value * 9/5) + 32
        elif to_unit == "kelvin":
            result = value + 273.15
    elif from_unit == "fahrenheit":
        if to_unit == "celsius":
            result = (value - 32) * 5/9
        elif to_unit == "kelvin":
            result = (value - 32) * 5/9 + 273.15
    elif from_unit == "kelvin":
        if to_unit == "celsius":
            result = value - 273.15
        elif to_unit == "fahrenheit":
            result = (value - 273.15) * 9/5 + 32

# Weight conversions
elif choice == "3":
    if from_unit == "kilograms":
        kg = value
    elif from_unit == "grams":
        kg = value / 1000
    elif from_unit == "pounds":
        kg = value * 0.453592
    else:
        kg = None

    if kg is not None:
        if to_unit == "kilograms":
            result = kg
        elif to_unit == "grams":
            result = kg * 1000
        elif to_unit == "pounds":
            result = kg / 0.453592

# Time conversions
elif choice == "4":
    if from_unit == "seconds":
        seconds = value
    elif from_unit == "minutes":
        seconds = value * 60
    elif from_unit == "hours":
        seconds = value * 3600
    else:
        seconds = None

    if seconds is not None:
        if to_unit == "seconds":
            result = seconds
        elif to_unit == "minutes":
            result = seconds / 60
        elif to_unit == "hours":
            result = seconds / 3600


if result is not None:
    print(f"{value} {from_unit} = {result:.2f} {to_unit}")
else:
    print("Invalid conversion units entered.")
```
