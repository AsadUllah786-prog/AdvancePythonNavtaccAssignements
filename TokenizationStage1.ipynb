```python
import re
```

```python
text = "Hello, world. This, is a test."
splittedtokens=re.split(r'([.,]|\s)',text)
tokens=[item for item in splittedtokens if item.strip()]
```

```python
print(tokens)
```

```python
text = "Hello, world. This, is a test. and i; ll?"
splite=re.split(r'([.,?!"";]|--|\s)',text)
print(splite)
tokens=[item for item in splite if item.strip()]
```

```python
print(tokens)
print(len(tokens))
```

```python
uniquetokens=sorted(set(tokens))
```

```python
print(len(uniquetokens))
```

```python
vocabdict={token:integer for integer, token in enumerate(uniquetokens)}
```

```python
print(vocabdict)
```
