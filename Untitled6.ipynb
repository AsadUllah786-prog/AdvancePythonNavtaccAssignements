```python
import pandas as pd
import numpy as np

# Create sample sales dataset again after reset
np.random.seed(42)

data = {
    "Date": pd.date_range(start="2024-01-01", periods=1000, freq="D"),
    "Region": np.random.choice(["North", "South", "East", "West"], 1000),
    "Product": np.random.choice(["Laptop", "Phone", "Tablet", "Headphones"], 1000),
    "Salesperson": np.random.choice(["Alice", "Bob", "Charlie", "David"], 1000),
    "Units Sold": np.random.randint(1, 20, 1000),
    "Unit Price": np.random.randint(100, 1000, 1000)
}
df = pd.DataFrame(data)
df["Revenue"] = df["Units Sold"] * df["Unit Price"]

# Save dataset to Excel for Dashboard
file_path = "C:/Users/User/OneDrive/Documents/excel/Sales_Dashboard_Data.xlsx"

df.to_excel(file_path, index=False)

file_path
```
