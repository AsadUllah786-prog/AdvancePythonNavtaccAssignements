Task 3


```python
# Input the sentence from user
sentence = input("Enter a sentence: ")
# Reverse the sentence
print(f"Reversed: {sentence[::-1]}")
# To count the vowels in the sentence
vowel = "aeiouAEIOU"
vowelCount = 0
for ch in sentence:
    if ch in vowel:
        vowelCount += 1
print(vowelCount)
```
