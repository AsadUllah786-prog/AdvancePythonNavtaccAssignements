```python
import requests
from bs4 import BeautifulSoup
import csv
import time

base_url = "https://remoteok.com/remote-python-jobs?page="

with open("python-jobs","w", newline="" ,encoding="UTF-8") as file:
    writer=csv.writer(file)
    writer.writerow(["Job Title","Company Name","Date Posted"])

    for page in range (1,6):
        url=base_url+str(page)
        response=requests.get(url,headers={"User-Agent": "Mozilla/5.0"})
        soup=BeautifulSoup(response.text,"html.parser")

        jobs=soup.find_all("tr",class_="job")

        for job in jobs:
            title_tag=job.find("h2",itemprop="title")
            company_tag = job.find("h3", itemprop="name")
            date_tag = job.find("time")

            if title_tag and company_tag and date_tag:
                title=title_tag.text.strip()
                companyName=company_tag.text.strip()
                date=date_tag.text.strip()

                writer.writerow([title,companyName,date])

        time.sleep(1)

print("all the data scraped successfully")
```
