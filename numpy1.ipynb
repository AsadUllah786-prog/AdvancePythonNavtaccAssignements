```python
data = np.array([[100, 200], [300, 400], [500, 600]])

mean = data.mean(axis=0)
std = data.std(axis=0)

normalized = (data - mean) / std
print(normalized)
```

```python
import numpy as np

scores=np.array([
    [74,81,92],
    [45,88,92],
    [80,70,90]
])

mean=scores.mean(axis=0)
std=scores.std(axis=0)

normalized=(scores-mean)/std
print(normalized)
```

```python
import numpy as np

# Simulated 4x4 grayscale image (pixel values 0-255)
image = np.array([
    [120, 135, 100, 90],
    [80, 150, 200, 175],
    [60, 65, 70, 85],
    [250, 240, 230, 220]
])

# Increase brightness by 30 (clipping to 255)
brighter_image = np.clip(image + 30, 0, 255)

print("Original Image:\n", image)
print("\nBrighter Image:\n", brighter_image)
```

```python
import numpy as np

# Sales data for 6 months (in 1000s)
sales = np.array([12, 15, 18, 22, 26, 31])

# Calculate monthly growth
growth = sales[1:] - sales[:-1]

# Percentage growth
percent_growth = (growth / sales[:-1]) * 100

print("Monthly Sales:", sales)
print("Monthly Growth:", growth)
print("Monthly Growth %:", np.round(percent_growth, 2))
```
