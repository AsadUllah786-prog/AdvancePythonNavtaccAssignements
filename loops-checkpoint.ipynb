```python
num=int(input("enter a number"))
original=num
reverse=0

while num>0:
    digit=num%10
    reverse=reverse*10+digit
    num//=10
if original==reverse:
    print("palindrome")
else:
    print("not a palindrome")
    
```

```python
text = input("Enter a string: ")
freq = {}
for char in text:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1
print(freq)
```
