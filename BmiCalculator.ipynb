# BMI CALCULATOR

```python
name=input("enter your name")
weight=int(input("enter your weight in pounds"))
height=int(input("enter your height in inches"))

BMI=(weight*703)/(height*height)

print(f"Asad your calculated BMI is {BMI}")
```

```python
if (BMI>0):
    if (BMI<18.5):
        print(f"{name} you are underweight")
    elif (BMI<24.5):
        print(f"{name} you are good to go")
    elif (BMI<34.5):
        print(f"{name} you are overweight focus on exercise")
    else:
        print(f"{name} you are not falling any of the category")
else:
    print("invalid inputs")
    
        
```
