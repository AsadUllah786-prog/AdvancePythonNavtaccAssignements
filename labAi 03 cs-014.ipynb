```python
def grade_calculate():
    percentage = float(input("Enter your percentage: "))
    if percentage >= 90:
        return f"your percentage is {percentage} and your grade is A"
    elif percentage >= 80:
        return f"your percentage is {percentage} and your grade is B"
    elif percentage >= 70:
        return f"your percentage is {percentage} and your grade is C"
    elif percentage >= 60:
        return f"your percentage is {percentage} and your grade is D"
    else:
        return "F"
print(grade_calculate())
```

```python
def percentage_calculate():
    obtainedScore=int(input("enter your obtained marks"))
    totalScore = int(input("Enter the total marks: "))
    percentage = (obtainedScore / totalScore) * 100
    return percentage
print(percentage_calculate())
```

```python
# Task 3
# Create a list of marks
marks = [78, 56, 89, 92, 67, 45]

# Function to calculate the average of marks
def average_marks(marks):
    return sum(marks) / len(marks)

# Calculate average
avg = average_marks(marks)
print("Average Marks:", avg)

# Find highest and lowest marks
highest = max(marks)
lowest = min(marks)

print("Highest Marks:", highest)
print("Lowest Marks:", lowest)



```

```python
# Task 4
# Add a new student's marks
marks=[92,89,85,78,67,56]
marks.append(85)
print("After adding new marks:", marks)

# Remove the lowest marks
marks.remove(min(marks))
print("After removing lowest marks:", marks)

# Sort the marks in descending order
marks.sort(reverse=True)
print("Marks in descending order:", marks)
```

```python
def get_top_student(students):
    # Find the student with the highest marks
    highest_mark = max(students.values())
    for name, mark in students.items():
        if mark == highest_mark:
            return name, mark  # return both name and marks
# Call the function
top_name, top_mark = get_top_student(students)
print("Top Student:", top_name)
print("Highest Marks:", top_mark)
```
