```python
from queue import Queue, LifoQueue, PriorityQueue

# Graph with edge costs
graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'D': 2, 'E': 12},
    'C': {'D': 5},
    'D': {'E': 3},
    'E': {}
}

# ---------- Breadth-First Search ----------
def bfs(start, goal):
    visited = []
    queue = Queue()
    queue.put([start])
    
    while not queue.empty():
        path = queue.get()
        node = path[-1]
        
        if node == goal:
            return path
        
        for neighbor in graph[node]:
            if neighbor not in path:
                new_path = list(path)
                new_path.append(neighbor)
                queue.put(new_path)
    return None

# ---------- Depth-First Search ----------
def dfs(start, goal):
    stack = LifoQueue()
    stack.put([start])
    
    while not stack.empty():
        path = stack.get()
        node = path[-1]
        
        if node == goal:
            return path
        
        for neighbor in graph[node]:
            if neighbor not in path:
                new_path = list(path)
                new_path.append(neighbor)
                stack.put(new_path)
    return None

# ---------- Uniform Cost Search ----------
def uniform_cost_search(start, goal):
    pq = PriorityQueue()
    pq.put((0, [start]))  # (cost, path)
    
    while not pq.empty():
        cost, path = pq.get()
        node = path[-1]
        
        if node == goal:
            return path, cost
        
        for neighbor, step_cost in graph[node].items():
            if neighbor not in path:
                new_path = list(path)
                new_path.append(neighbor)
                pq.put((cost + step_cost, new_path))
    return None, None

# ---------- Run All ----------
print("BFS Path:", bfs('A', 'E'))
print("DFS Path:", dfs('A', 'E'))
ucs_path, ucs_cost = uniform_cost_search('A', 'E')
print("UCS Path:", ucs_path, "Cost:", ucs_cost)
```
