```python
import requests
```

```python
URL="http://quotes.toscrape.com"
```

```python
response=requests.get(URL)
```

```python
print(response.status_code)
```

```python
print(response.text[:500])
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(response.text,"html.parser")
```

```python
quotes=soup.find_all("div",class_="quote")
```

```python
print(f"Found {len(quotes)} quotes on page")
```

```python
first_quote=quotes[0]    
```

```python
text=first_quote.find("span", class_="text").text
print(text)
```

```python
author=first_quote.find("small", class_="author").text
print(author)
```

```python
tags=[tag.text for tag in first_quote.find_all("a",class_="tag")]
```

```python
print(tags)
```

```python
all_data = []
for quote in quotes:
    text=quote.find("span",class_="text").text
    author=quote.find("small",class_="author").text
    tags=[tag.text for tag in quote.find_all("a",class_="tag")]

    all_data.append([text,author," ".join(tags)])

print(all_data[:2])
    
```

```python
import  csv
import os

desktop_path = os.path.join(os.path.expanduser("~"), r"C:\Users\User", "quotes.csv")
with open(desktop_path,"w" ,newline="", encoding="UTF-8") as file:
    writer=csv.writer(file)
    writer.writerow(["QUOTES","AUTHOR","TAGS"])
    writer.writerows(all_data)


print("file saved")

```
