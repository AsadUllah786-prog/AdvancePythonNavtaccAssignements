```python
import pandas as pd
import random
import os

questions = pd.read_csv("Questionaire.csv", sep=",")
questions.columns = questions.columns.str.strip() 


if os.path.exists("student_score.csv"):
    scores = pd.read_csv("student_score.csv")
else:
    scores = pd.DataFrame(columns=["Name", "Score"])

name = input("Enter your name: ").strip()


if name in scores["Name"].values:
    print("You have already taken the quiz. One attempt only!")
else:
    quiz_questions = questions.sample(n=10) 
    score = 0
    i = 0

    
    while i < len(quiz_questions):
        q = quiz_questions.iloc[i]

        print(f"\n{q['Question']}")
        print(f"A. {q['Option A']}")
        print(f"B. {q['Option B']}")
        print(f"C. {q['Option C']}")
        print(f"D. {q['Option D']}")

        answer = input("Your answer (A/B/C/D): ").strip().upper()
        correct_letter = q['Right Answer']
        correct_value = q[f"Option {correct_letter}"]
        if answer == q['Right Answer']:
            score += 1
            i += 1
        else:
            print(f"❌ Wrong answer! Game Over. the correct answer is {correct_value}")
           
            scores.loc[len(scores)] = [name, score]
            scores.to_csv("student_score.csv", index=False)
            break  # end instantly

    # If loop finishes, student got all correct
    print(f"\n✅ Quiz completed! Your score: {score}")
    scores.loc[len(scores)] = [name, score]
    scores.to_csv("student_score.csv", index=False)
```
