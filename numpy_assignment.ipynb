# Task 3: Basic Array Operations


```python
A=[1,2,3,4,5]
B=[5,4,3,2,1]

firstarray=np.array(A)
secondarray=np.array(B)
```

```python
Addition=(firstarray+secondarray)
Subtracction=(firstarray-secondarray)
Multiplication=(firstarray*secondarray)
Division=(firstarray/secondarray)
```

```python
print(Addition)
print(Subtracction)
print(Multiplication)
print(Division)
```

# Task 8: Matrix Operations

```python
matrix1=np.matrix([[1,2,3],[4,5,6]])
matrix2=np.matrix([[1,2,3],[4,5,6]])
```

```python
addition=matrix1+matrix2
subtraction=matrix1-matrix2
multiplication=np.multiply(matrix1,matrix2)
```

```python
print(addition)
print(subtraction)
print(multiplication)
```

#  Task 10: Practical Problem

```python
marks = np.array([70, 85, 90, 60, 75, 88, 92, 55, 80, 77])


print("Highest marks:", np.max(marks))


print("Lowest marks:", np.min(marks))


avg = np.mean(marks)
print("Average marks:", avg)


print("Above average:", marks[marks > avg])
```
