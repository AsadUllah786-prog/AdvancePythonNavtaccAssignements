```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Convert all to numeric
for c in pima.columns:
    pima[c] = pd.to_numeric(pima[c], errors='coerce')

# Remove rows where LABEL is NaN
pima = pima.dropna(subset=['label'])

# Split features & target
X = pima[['pregnant', 'insulin', 'bmi', 'age', 'glucose', 'bp', 'pedigree']]
y = pima['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=16)

# Pipeline: impute + scale + logistic regression
pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
    ('model', LogisticRegression(max_iter=1000))
])

pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
```

```python
# import the metrics class
from sklearn import metrics

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python



import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

# ----------------------------
# 1. Convert all columns to numeric
# ----------------------------
for c in pima.columns:
    pima[c] = pd.to_numeric(pima[c], errors='coerce')

# ----------------------------
# 2. Remove rows where label is NaN
# ----------------------------
pima = pima.dropna(subset=['label'])

# ----------------------------
# 3. Split features & target
# ----------------------------
feature_cols = ['pregnant', 'insulin', 'bmi', 'age', 'glucose', 'bp', 'pedigree']
X = pima[feature_cols]
y = pima['label']

# ----------------------------
# 4. Train/test split
# ----------------------------
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=16)

# ----------------------------
# 5. Pipeline: imputer + scaler + logistic regression
# ----------------------------
pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
    ('model', LogisticRegression(max_iter=1000))
])

# Train the model
pipe.fit(X_train, y_train)

# ----------------------------
# 6. Predictions & probabilities
# ----------------------------
y_pred = pipe.predict(X_test)
y_pred_proba = pipe.predict_proba(X_test)[:, 1]

# ----------------------------
# 7. ROC curve and AUC
# ----------------------------
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)

plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, label="Logistic Regression, AUC="+str(round(auc,3)))
plt.plot([0,1], [0,1], 'k--')  # random line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc=4)
plt.show()

# ----------------------------
# 8. Optional: Accuracy & Classification Report
# ----------------------------
print("Accuracy:", pipe.score(X_test, y_test))
print(metrics.classification_report(y_test, y_pred))
```
