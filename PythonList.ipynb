```python
orders = [
    (1001, "Ali", "Lahore", "Laptop", 1, 120000),
    (1002, "Sara", "Karachi", "Mobile", 2, 40000),
    (1003, "Ahmed", "Lahore", "Tablet", 1, 60000),
    (1004, "Ali", "Islamabad", "Mobile", 1, 40000),
    (1005, "Zara", "Karachi", "Laptop", 1, 120000),
    (1006, "Ahmed", "Lahore", "Mobile", 3, 40000),
]
```

```python
# Data stored as tuple but it is a combinations or group in list

# tuple immutable duplicates index based
# list mutable  duplicates  index based 

totalRevenue=sum(order[4]*order[5] for order in orders)
print(totalRevenue)



avergeamount=totalRevenue/(len(orders))
print(avergeamount)
```

```python
uniqueCities={order[2] for order in orders}
print(uniqueCities)
```

```python
uniqueNames={name[1] for name in orders}
print(uniqueNames)
```

```python
LAHORE_ORDERS=[order for order in orders if order[2]=="Lahore"]
print(LAHORE_ORDERS)
```

```python
customer_sales = {}

for order in orders:
    name = order[1]
    revenue = order[4] * order[5]
    customer_sales[name] = customer_sales.get(name, 0) + revenue

print(customer_sales)
```

```python
UNIQUE_PRODUCTS={order[3] for order in orders}
print(UNIQUE_PRODUCTS)
```
