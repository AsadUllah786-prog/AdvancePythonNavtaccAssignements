```python
print("Marks, Grade, and Average Calculation System")

print("this system calculates the average and grade as a 100 marks numbers per subject")


num_subjects = int(input("Enter the number of subjects: "))

total = 0


for i in range(num_subjects):
    marks = float(input(f"Enter marks for subject {i+1}: "))
    total += marks


average = total / num_subjects


if average >= 80:
    grade = "A"
elif average >= 70:
    grade = "B"
elif average >= 60:
    grade = "C"
elif average >= 50:
    grade = "D"
else:
    grade = "F"


print("\n--- Results ---")
print(f"Total Marks = {total}")
print(f"Average Marks = {average:.2f}")
print(f"Grade = {grade}")
```
