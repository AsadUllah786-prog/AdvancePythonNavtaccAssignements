```python
sales = [100, 102, 98, 101, 99, 140]

sales_series = pd.Series(sales)

mean = sales_series.mean()
variance = sales_series.var()
std_dev = sales_series.std()

print("Mean:", mean)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
plt.figure()
plt.hist(sales, bins=5)

plt.axvline(mean)
plt.axvline(mean + std_dev)
plt.axvline(mean - std_dev)

plt.title("Sales Distribution with Standard Deviation")
plt.show()
```
