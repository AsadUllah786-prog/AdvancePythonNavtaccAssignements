```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
def createAndTrainModel(df,features,target):
    X = df[features]
    Y = df[target]

    xtrain, xtest, ytrain, y_test = train_test_split(
        X, Y, test_size=0.2, random_state=42
    )

    
    model = LinearRegression()
    model.fit(xtrain, ytrain)

    y_pred = model.predict(xtest)

    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print(f"Mean Squared Error: {mse:.2f}")
    print(f"R² Score: {r2:.2f}")

    print(f"Coefficients: {dict(zip(features,model.coef_))}")
    print(f"Coffecient:{dict(zip(features,model.coef_))}")
    print(f"Intercept: {model.intercept_:.2f}")

    return model, y_pred

features = ["TRIP_SECONDS", "TRIP_MILES", "TIP_RATE"]
target = "FARE"
createAndTrainModel(df, features, target)
```
