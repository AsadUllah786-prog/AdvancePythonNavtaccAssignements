```python
import random as rd
```

```python
rd.seed(42)
```

```python
number=rd.randint(1,50)
```

```python
attempts=0
```

```python
while True:
    userguess=int(input("please enter a number in the range between (1,50)"))
    if userguess>number:
        print("too high")
        attempts+=1
    elif userguess<number:
        print("too low")
        attempts+=1
    elif userguess==number:
        print(f"correct guess in {attempts} attempts")
        break
    else:
        print("invalid guess")
```
