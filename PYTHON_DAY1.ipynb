```python
raw_orders = [
    (101, "Laptop", 1, 1200),
    (102, "Mouse", 2, 25),
    (101, "Keyboard", 1, 75),
    (103, "Monitor", 1, 300),
    (102, "Mouse Pad", 1, 10),
    (104, "Laptop", 1, 1200),
    (101, "Laptop", 1, 1200) # Duplicate entry for order 101, same product
]
```

```python
print(raw_orders)
```

```python
unique_orders=set(raw_orders)
```

```python
product_sales={}

for order in raw_orders:
    product_name=order[1]
    quantity=order[2]
    price=order[3]
    total_item_price = quantity * price

    product_sales[product_name] = product_sales.get(product_name, 0) + total_item_price

```

```python
print(product_sales)
```
