```python
import pandas as pd
import random
import os

# Load questions (adjust sep depending on your file: "," or "\t")
questions = pd.read_csv("Questionaire.csv", sep=",")
questions.columns = questions.columns.str.strip()   # clean headers

# Load or create student score file
if os.path.exists("student_score.csv"):
    scores = pd.read_csv("student_score.csv")
else:
    scores = pd.DataFrame(columns=["Name", "Score"])

# Ask for student name
name = input("Enter your name: ").strip()

# Check if already attempted
if name in scores["Name"].values:
    print("⚠️ You have already taken the quiz. One attempt only!")
else:
    quiz_questions = questions.sample(n=7)  # pick 7 random questions
    score = 0
    i = 0

    # While loop goes through questions
    while i < len(quiz_questions):
        q = quiz_questions.iloc[i]

        print(f"\n{q['Question']}")
        print(f"A. {q['Option A']}")
        print(f"B. {q['Option B']}")
        print(f"C. {q['Option C']}")
        print(f"D. {q['Option D']}")

        answer = input("Your answer (A/B/C/D): ").strip().upper()
        if answer == q['Right Answer']:
            score += 1
            i += 1
        else:
            print("❌ Wrong answer! Game Over.")
            scores.loc[len(scores)] = [name, score]
            scores.to_csv("student_score.csv", index=False)
            break  # end instantly

    # If loop finishes, student got all correct
    print(f"\n✅ Quiz completed! Your score: {score}")
    scores.loc[len(scores)] = [name, score]
    scores.to_csv("student_score.csv", index=False)
```

```python
import pandas as pd
import random
import os
import tkinter as tk
from tkinter import messagebox


questions = pd.read_csv("Questionaire.csv")
questions.columns = questions.columns.str.strip()

if os.path.exists("student_score.csv"):
    scores = pd.read_csv("student_score.csv")
else:
    scores = pd.DataFrame(columns=["Name", "Score"])


quiz_questions = None
score = 0
i = 0
name = ""
time_left = 60


def start_quiz():
    global name, quiz_questions, score, i, time_left

    name = name_entry.get().strip()

    if name == "":
        messagebox.showerror("Error", "Enter name")
        return

    if name in scores["Name"].values:
        messagebox.showinfo("Info", "One attempt only")
        return

    quiz_questions = questions.sample(n=10)

    score = 0
    i = 0
    time_left = 60

    login_frame.pack_forget()
    quiz_frame.pack()

    show_question()
    timer()


def show_question():
    global i

    if i >= len(quiz_questions):
        finish_quiz()
        return

    q = quiz_questions.iloc[i]

    q_no_label.config(
        text=f"Question {i+1} / 10"
    )

    question_label.config(
        text=q["Question"]
    )

    var.set("")

    r1.config(text=q["Option A"], value="A")
    r2.config(text=q["Option B"], value="B")
    r3.config(text=q["Option C"], value="C")
    r4.config(text=q["Option D"], value="D")


def next_question():
    global i, score

    ans = var.get()

    if ans == "":
        messagebox.showwarning("Warning", "Select answer")
        return

    q = quiz_questions.iloc[i]

    if ans == q["Right Answer"]:
        score += 1
    else:
        correct = q[f"Option {q['Right Answer']}"]
        messagebox.showinfo(
            "Wrong",
            f"Correct = {correct}"
        )

    i += 1
    show_question()


def timer():
    global time_left

    if time_left <= 0:
        finish_quiz()
        return

    timer_label.config(
        text=f"Time: {time_left}"
    )

    time_left -= 1
    root.after(1000, timer)


def finish_quiz():

    scores.loc[len(scores)] = [name, score]
    scores.to_csv("student_score.csv", index=False)

    quiz_frame.pack_forget()
    result_frame.pack()

    result_label.config(
        text=f"Exam Finished\nScore = {score}"
    )


def show_scoreboard():

    top = tk.Toplevel(root)
    top.title("Scoreboard")

    t = tk.Text(top, width=30, height=15)
    t.pack()

    for _, row in scores.iterrows():
        t.insert(
            tk.END,
            f"{row['Name']}   {row['Score']}\n"
        )


root = tk.Tk()
root.title("Exam Software")
root.geometry("600x500")


login_frame = tk.Frame(root)
quiz_frame = tk.Frame(root)
result_frame = tk.Frame(root)


login_frame.pack()


tk.Label(
    login_frame,
    text="Exam System",
    font=20
).pack(pady=10)

tk.Label(
    login_frame,
    text="Enter Name"
).pack()

name_entry = tk.Entry(login_frame)
name_entry.pack()

tk.Button(
    login_frame,
    text="Start Exam",
    bg="green",
    fg="white",
    command=start_quiz
).pack(pady=5)

tk.Button(
    login_frame,
    text="Scoreboard",
    command=show_scoreboard
).pack()



top_frame = tk.Frame(quiz_frame)
top_frame.pack()

q_no_label = tk.Label(top_frame, text="")
q_no_label.pack(side="left", padx=10)

timer_label = tk.Label(top_frame, text="")
timer_label.pack(side="right", padx=10)



question_label = tk.Label(
    quiz_frame,
    text="",
    wraplength=500,
    font=12
)
question_label.pack(pady=20)


var = tk.StringVar()

r1 = tk.Radiobutton(quiz_frame, text="", variable=var, value="A")
r1.pack(anchor="w")

r2 = tk.Radiobutton(quiz_frame, text="", variable=var, value="B")
r2.pack(anchor="w")

r3 = tk.Radiobutton(quiz_frame, text="", variable=var, value="C")
r3.pack(anchor="w")

r4 = tk.Radiobutton(quiz_frame, text="", variable=var, value="D")
r4.pack(anchor="w")


tk.Button(
    quiz_frame,
    text="Next",
    bg="blue",
    fg="white",
    command=next_question
).pack(pady=10)



result_label = tk.Label(
    result_frame,
    text="",
    font=16
)
result_label.pack(pady=20)

tk.Button(
    result_frame,
    text="Close",
    command=root.destroy
).pack()


root.mainloop()
```
