```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist=[]

for x in fruits:
    if (x!="apple" and x!="banana"):
        newlist.append(x)
print(newlist)   

newlist1 = [x.upper() for x in fruits]
print(newlist1)
```
