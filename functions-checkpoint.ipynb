```python
def tri_recursion(k):
  if(k > 0):
    result = k + tri_recursion(k - 1)
    print(result)
  else:
    result = 0
  return result

print("Recursion Example Results:")
tri_recursion(6)
```

```python


def check_even_odd(n):
    even=0
    odd=0
    for i in range(n):
        if i%2==0:
            even+=i
        else:
            odd+=i

    return (even,odd)
        
check_even_odd(10)        
    
```

```python
def even_or_odd(start, end):
    for i in range(start, end+1):
        if i % 2 == 0:
            print(f"{i} is Even")
        else:
            print(f"{i} is Odd")

even_or_odd(1, 5)
```
