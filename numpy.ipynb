```python
images = np.random.randint(0, 256, (3, 4, 4))  # shape (3,4,4)

# Compute mean & std per image
mean = images.mean(axis=(1, 2))  # shape (3,)
std = images.std(axis=(1, 2))    # shape (3,)

# Expand dims to match (3,4,4)
mean = np.expand_dims(mean, axis=(1, 2))
std = np.expand_dims(std, axis=(1, 2))

# Broadcasting mean & std over each pixel
normalized_images = (images - mean) / std

print("Original shape:", images.shape)
print("Normalized shape:", normalized_images.shape)
```
