```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sales=np.random.normal(200,20,10000)

prob_below70=np.sum(sales<170)/len(sales)


print(prob_below70)

plt.hist(sales, bins=50)
plt.axvline(170)
plt.show()
```

```python
# 6 % are chances that sales are below 170

prob_greater70=1-prob_below70

print(prob_greater70)
```
