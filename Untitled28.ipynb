```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
numeric_cols


def iqr_summary(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Outlier mask
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    outlier_count = outliers.shape[0]

    return {
        "Q1": Q1,
        "Q3": Q3,
        "IQR": IQR,
        "lower_bound": lower_bound,
        "upper_bound": upper_bound,
        "outliers": outlier_count
    }
```

```python
iqr_results = {}

for col in numeric_cols:
    iqr_results[col] = iqr_summary(df, col)
```

```python
for col, result in iqr_results.items():
    print(f"\n===== {col} =====")
    print(f"Q1:           {result['Q1']}")
    print(f"Q3:           {result['Q3']}")
    print(f"IQR:          {result['IQR']}")
    print(f"Lower Bound:  {result['lower_bound']}")
    print(f"Upper Bound:  {result['upper_bound']}")
    print(f"Outliers:     {result['outliers']}")
```
