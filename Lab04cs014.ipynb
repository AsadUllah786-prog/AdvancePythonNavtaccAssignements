```python
movies = {
    "X-Men First Class": ["Jennifer Lawrence", "Kevin Bacon"],
    "Apollo 13": ["Tom Hanks", "Kevin Bacon"],
    "Titanic": ["Leonardo DiCaprio", "Kate Winslet"]
}

movie_name = "X-Men First Class"
actors = movies[movie_name]

print(actors)

for a in actors:
    print(a)
    for b in actors:
        print(b)
        if a != b:
            print(f"{a} worked with {b} in {movie_name}")
```

```python
movies = {
    "X-Men First Class": ["Jennifer Lawrence", "Kevin Bacon"],
    "Apollo 13": ["Tom Hanks", "Kevin Bacon"],
    "Titanic": ["Leonardo DiCaprio", "Kate Winslet"]
}

connections = {}

for movie, stars in movies.items():
    for actor in stars:
        # If this actor isn't in the dictionary yet, add them with an empty set
        if actor not in connections:
            connections[actor] = set()
        # Add all co-stars (everyone else in the same movie)
        for co_star in stars:
            if co_star != actor:
                connections[actor].add(co_star)

print("Jennifer Lawrence’s connections:", connections["Jennifer Lawrence"])

print(connections)
```
