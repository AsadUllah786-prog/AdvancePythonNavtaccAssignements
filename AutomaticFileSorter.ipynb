```python
import os
import shutil
from time import sleep

WATCH_FOLDER = r"C:\Users\User\OneDrive\Documents\FileSorter"

FILE_TYPES = {
    "Images": [".png", ".jpg", ".jpeg", ".gif"],
    "Documents": [".pdf", ".docx", ".txt", ".pptx",".csv"],
    "Videos": [".mp4", ".mkv", ".mov"],
    "Music": [".mp3", ".wav"],
    "Archives": [".zip", ".rar", ".7z"],
    "Code": [".py", ".cpp", ".js", ".html", ".css"]
}

def create_folders():
    for folder in FILE_TYPES:
        path = os.path.join(WATCH_FOLDER, folder)
        os.makedirs(path, exist_ok=True)

def sort_files():
    for filename in os.listdir(WATCH_FOLDER):
        file_path = os.path.join(WATCH_FOLDER, filename)

        # Skip directories
        if os.path.isdir(file_path):
            continue

        # Get extension
        _, ext = os.path.splitext(filename)
        ext = ext.lower()

        # Match extension
        for folder, extensions in FILE_TYPES.items():
            if ext in extensions:
                dest = os.path.join(WATCH_FOLDER, folder, filename)

                # Avoid overwriting files
                if not os.path.exists(dest):
                    shutil.move(file_path, dest)
                    print(f"Moved: {filename} → {folder}")
                break

def run_sorter():
    print("Auto File Sorter Running...")
    create_folders()

    while True:
        sort_files()
        sleep(2)

run_sorter()
```
