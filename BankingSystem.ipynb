```python
bank = {}

while True:
    print("\n1. Register Account")
    print("2. Access Account")
    print("3. Exit")

    choice = input("Enter choice: ")

  
    if choice == "1":
        username = input("Enter username: ")

        if username in bank:
            print("User already exists. Access denied.")
        else:
            age = int(input("Enter age: "))
            
            
            bank[0] = {
                "Name":[username],
                "age": [age],
                "balance": 5000
            }

            print("Account created with 5000 initial balance.")

    elif choice == "2":
        username = input("Enter username: ")

        if username not in bank:
            print("User does not exist.")
        else:
            while True:
                print("\n1. Withdraw")
                print("2. Deposit")
                print("3. Check Balance")
                print("4. Back")

                option = input("Select option: ")

                
                if option == "1":
                    amount = float(input("Enter amount to withdraw: "))
                    
                    if amount <= bank[username]["balance"]:
                        bank[username]["balance"] -= amount
                        print("Withdrawal successful.")
                    else:
                        print("Insufficient balance.")

                
                elif option == "2":
                    amount = float(input("Enter amount to deposit: "))
                    bank[username]["balance"] += amount
                    print("Deposit successful.")

                
                elif option == "3":
                    print("Current Balance:", bank[username]["balance"])

            
                elif option == "4":
                    break

                else:
                    print("Invalid option.")

   
    elif choice == "3":
        print("System Closed.")
        break

    else:
        print("Invalid choice.")
```

```python
import csv
import os

file_name = "bank.csv"

users = []



if os.path.exists(file_name):

    with open(file_name, "r", newline="") as f:
        reader = csv.DictReader(f)

        for row in reader:
            row["age"] = int(row["age"])
            row["balance"] = int(row["balance"])
            users.append(row)


while True:

    print("\n1 Create account")
    print("2 Deposit")
    print("3 Withdraw")
    print("4 Show users")
    print("5 Exit")

    choice = input("Enter choice: ")


    
    if choice == "1":

        name = input("Enter name: ")
        age = int(input("Enter age: "))

        user = {
            "name": name,
            "age": age,
            "balance": 5000
        }

        users.append(user)

        print("Account created")


    
    elif choice == "2":

        name = input("Enter name: ")

        for u in users:

            if u["name"] == name:

                amount = int(input("Enter amount: "))
                u["balance"] += amount

                print("Deposit done")



    elif choice == "3":

        name = input("Enter name: ")

        for u in users:

            if u["name"] == name:

                amount = int(input("Enter amount: "))

                if amount <= u["balance"]:
                    u["balance"] -= amount
                    print("Withdraw done")

                else:
                    print("Not enough balance")


 
    elif choice == "4":

        for u in users:
            print(u)


    elif choice == "5":
        break


  
    with open(file_name, "w", newline="") as f:

        fieldnames = ["name", "age", "balance"]

        writer = csv.DictWriter(f, fieldnames=fieldnames)

        writer.writeheader()

        for u in users:
            writer.writerow(u)
```
